```python
tlefile=r'C:\Users\jinni\Hardware_Tracking_Software\gps-ops.txt'
TLEfile=open(tlefile, 'rt')

SatCons = list()
line0=''
i=0
try:
    while True:
        line0=TLEfile.readline()
        line1=TLEfile.readline()
        line2=TLEfile.readline()
        Satellite_i = fio.Satellite(line0, line1, line2)
        print(Satellite_i)
        SatCons.append(Satellite_i)
        i+=1
except ValueError:
    print("Read in", len(SatCons), "satellites")
    TLEfile.close()
```

```python
# Open the TLE file safely
try:
    with open(tlefile, 'rt') as TLEfile:
        SatCons = []  # List to store satellite objects
        i = 0

        while True:
            line0 = TLEfile.readline().strip()  # Read first line (satellite name)
            line1 = TLEfile.readline().strip()  # Read second line (TLE line 1)
            line2 = TLEfile.readline().strip()  # Read third line (TLE line 2)

            # ✅ Ensure all lines are valid before processing
            if not line0 or not line1 or not line2:
                break  # Stop reading when reaching the end of the file

            # ✅ Create Satellite object and store in list
            try:
                Satellite_i = fio.Satellite(line0, line1, line2)  # Ensure fio.Satellite exists
                print(Satellite_i)
                SatCons.append(Satellite_i)
                i += 1
            except Exception as e:
                print(f"Error processing satellite {line0}: {e}")  # Print error without breaking

except FileNotFoundError:
    print(f"Error: TLE file not found at {tlefile}")

except Exception as e:
    print(f"Unexpected error: {e}")

finally:
    print(f"✅ Successfully read {len(SatCons)} satellites.")
```
